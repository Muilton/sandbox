## Параметрические критерии
Чтобы проверять гипотезы с применением критериев Стьюдента, необходимо чтобы исследуемые выборки не сильно отличались от нормального распределения. Для этого построим Q-Q график и проверим выборки с помощью критерия Шапиро-Уилка. <br/>
<br/>
*Гипотеза* <br/>
H0: Значения признака Races распределены нормально <br/>
H1: Не нормально
<br/>
По результатам проведенных тестов мы во всех случаях отклоняем нулевую гипотезу, следовательно, признаки распределены не нормально.


```python
ax1 = plt.subplot2grid(shape=(2,6), loc=(0,0), colspan=2)
ax2 = plt.subplot2grid((2,6), (0,2), colspan=2)
ax3 = plt.subplot2grid((2,6), (0,4), colspan=2)
ax4 = plt.subplot2grid((2,6), (1,1), colspan=2)
ax5 = plt.subplot2grid((2,6), (1,3), colspan=2)

plots = [ax1, ax2, ax3, ax4, ax5]

for i in range(1, 6):
    shapiro_races = scipy.stats.shapiro(df[df.Transport == i].Races)
    print(f"Shapiro normally test. Races. Transport={i}.  W-statistic: {shapiro_races[0]}, p-value: {shapiro_races[1]}")
    scipy.stats.probplot(df[df.Transport == i].Races, dist='norm', plot=plots[i-1]);
```

*Гипотеза* <br/>
H0: Значения признака mean_fuel распределены нормально <br/>
H1: Не нормально
<br/>
По результатам проведенных тестов мы во всех случаях отклоняем нулевую гипотезу, следовательно, признаки распределены не нормально.

```python
ax1 = plt.subplot2grid(shape=(2,6), loc=(0,0), colspan=2)
ax2 = plt.subplot2grid((2,6), (0,2), colspan=2)
ax3 = plt.subplot2grid((2,6), (0,4), colspan=2)
ax4 = plt.subplot2grid((2,6), (1,1), colspan=2)
ax5 = plt.subplot2grid((2,6), (1,3), colspan=2)

plots = [ax1, ax2, ax3, ax4, ax5]

for i in range(1, 6):
    shapiro_races = scipy.stats.shapiro(df[df.Transport == i].mean_fuel)
    print(f"Shapiro normally test. Races. Transport={i}.  W-statistic: {shapiro_races[0]}, p-value: {shapiro_races[1]}")
    scipy.stats.probplot(df[df.Transport == i].mean_fuel, dist='norm', plot=plots[i-1]);
```

Увы, оценка критерием Шапиро-Уилка подвержена неточности при наличие в данных выбросов (экстримальных отклонений от среднего). Исключим из выборок все значения, превышающие 95-персентиль и меньше 5-персентиля. <br/>
На новых данных построим Q-Q график и проверим на нормальность тем же методом.

```python
ax1 = plt.subplot2grid(shape=(2,6), loc=(0,0), colspan=2)
ax2 = plt.subplot2grid((2,6), (0,2), colspan=2)
ax3 = plt.subplot2grid((2,6), (0,4), colspan=2)
ax4 = plt.subplot2grid((2,6), (1,1), colspan=2)
ax5 = plt.subplot2grid((2,6), (1,3), colspan=2)

plots = [ax1, ax2, ax3, ax4, ax5]

for i in range(1, 6):
    q1 = df[df.Transport == i].mean_fuel.quantile(0.05) 
    q2 = df[df.Transport == i].mean_fuel.quantile(0.95) 
    
    shapiro_mean_fuel = scipy.stats.shapiro(df[(df.Transport == i) & (df.mean_fuel > q1) & (df.mean_fuel < q2)].mean_fuel)
    print(f"Shapiro normally test. mean_fuel. Transport={i}.  W-statistic: {shapiro_mean_fuel[0]}, p-value: {shapiro_mean_fuel[1]}")
    scipy.stats.probplot(df[(df.Transport == i) & (df.mean_fuel > q1) & (df.mean_fuel < q2)].mean_fuel, dist='norm', plot=plots[i-1]);
```

Даже после очистки данных, мы не можем считать наши выборки распределенными нормально. Также кретерий Колмогорова-Смирнова это подтвержает (не рассписывал). Следовательно, использование параметрических методов оценки среднего (в частности, критерий Стьюдента) мы не можем использовать.

## Непараметрические критерии

Просто построив 95% интервал среднего расхода топлива на рейс можно отличать некоторые типы грузовиков.

```python
colors = ['red', 'green', 'blue', 'black', 'orange']
plt.rcParams["figure.figsize"] = [10, 4]
for i in range(1, 6):
    s, f = zconfint(df[df.Transport == i].mean_fuel)
    print(f'Transport {i}. Среднее: {df[df.Transport == i].mean_fuel.mean()}. 95% интервал:', zconfint(df[df.Transport == i].mean_fuel))
    plt.axhline(y=i, xmin=s/100, xmax=f/100, c=colors[i-1])
```

Сделаем тоже самое по кол-ву рейсов.

```python
colors = ['red', 'green', 'blue', 'black', 'orange']
plt.rcParams["figure.figsize"] = [10, 4]
for i in range(1, 6):
    s, f = zconfint(df[df.Transport == i].Races)
    print(f'Transport {i}. Среднее: {df[df.Transport == i].mean_fuel.mean()}. 95% интервал:', zconfint(df[df.Transport == i].Races))
    plt.axhline(y=i, xmin=s/100, xmax=f/100, c=colors[i-1])
```

## Корреляция Races и Fuel

1) Посчитаем коэффициент корреляции Пирсона
2) Используем критерий Стьюдента для проверки гипотезы о зависимости данных признаков.<br/>
<br/>
*Гипотеза*<br/>
H0: Коэффициент корреляции равен 0 <br/>
H1: Не равен 0


```python
df[['Races', 'Fuel']].corr()
```

```python
# расчет доверительного интервала коэфф. корреляции

def r_to_z(r):
    return math.log((1 + r) / (1 - r)) / 2.0

def z_to_r(z):
    e = math.exp(2 * z)
    return((e - 1) / (e + 1))

def r_confidence_interval(r, alpha, n):
    z = r_to_z(r)
    se = 1.0 / math.sqrt(n - 3)
    z_crit = scipy.stats.norm.ppf(1 - alpha/2)

    lo = z - z_crit * se
    hi = z + z_crit * se

    # Return a sequence
    return (z_to_r(lo), z_to_r(hi))

```

```python
r, p = scipy.stats.pearsonr(df.Races, df.Fuel)
print(f"Для признаков Races и Fuel коэффициент корреляции Пирсона: {r}. P-value равно {p}")
print("95% Доверительный интервал коэффициента корреляции:", r_confidence_interval(r, 0.05, df.shape[0]))
```

Исходя из полученный рассчетов, мы с уверенность 95% отклоняем нулевую гипотезу (r == 0). Построенный доверительный интервал не включает в себя 0. <br/>
Следовательно, мы можем утверждать, что Races и Fuel зависимы.
