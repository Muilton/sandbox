```python
df = pd.read_table('pines.txt')
```

```python
df
```

```python
maps = scipy.stats.binned_statistic_2d(df.sn, df.we, values=df.sn, statistic='count', bins=[5, 5])[0]
maps.mean()
```

```python
maps = maps.reshape(1, 25)[0].astype('int')

observed_freq = np.bincount(maps)
expected_freq = [len(maps) * scipy.stats.poisson.pmf(x, maps.mean()) for x in range(min(maps), max(maps)+1)]

scipy.stats.chisquare(observed_freq, expected_freq, ddof=1)
```
